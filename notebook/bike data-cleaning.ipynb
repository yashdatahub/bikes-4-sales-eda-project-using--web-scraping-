## Extract the Columns from the containers which needed

| Field      | Regex                           |
|               |                                    |
| Bike Name  | `^(.*?),\s*\d{4}`               |
| Model Year | `,\s*(\d{4})\s*ModelLocation:`  |
| Location   | `ModelLocation:\s*(.*?)\s*Rs\.` |
| Price      | `Rs\.\s*([\d,]+)`               |


```python
import pandas as pd
import re

bike_name = []
model_year = []
location = []
price_list = [] 
finance =[]

raw_data = pd.read_csv(r"C:/Users/vinee/Innomatics- Data Analysis/Project/EDA_Data_Scraping/Bikes_data.csv")
text_data = raw_data.iloc[:, 0].astype(str)

for entry in text_data:

    # Bike Name
    name = re.search(r"^(.*?),\s*\d{4}", entry)
    bike_name.append(name.group(1).strip() if name else None)

    # Model Year
    year = re.search(r",\s*(\d{4})", entry)
    model_year.append(year.group(1) if year else None)

    # Location
    loc = re.search(r"ModelLocation:\s*(.*?)\s*Rs\.", entry)
    location.append(loc.group(1).strip().rstrip(",") if loc else None)

    # Price
    price_match = re.search(r"Rs\.\s*([\d,]+)", entry)
    price_list.append(price_match.group(1).replace(",", "") if price_match else None)

    # finance
    # Check Finance Available
    if "Finance Available" in entry:
        finance.append("Yes")
    else:
        finance.append("No")
        
df = pd.DataFrame({
    "bike_name": bike_name,
    "model_year": model_year,
    "location": location,
    "price(RS)": price_list,
    "finance(YES/NO)" : finance
})
```
